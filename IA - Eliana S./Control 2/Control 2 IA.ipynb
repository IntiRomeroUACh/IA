<h3><b>1. i) Agrupamiento Jerarquico</h3>

Descripción:

Método de aprendizaje no supervizado en el cual, se toma una conjunto de datos de entrada y se agrupan en una jerarquita de arboles anidados. Existen 2 enfoque principales en esta metodología, agrupamiento de arriba-abajo y agrupamiento de abajo-arriba. Ambos toman el conjunto de datos y los agrupan.

Ventajas:
Puede ser utilizada cualquier medida de distancia, no solo la euclideana.

Desventajas: Es dificil de determinar la calidad de los agrupamientos formados.

Ejemplos de aplicación:
Si se tienen datos con características similares y se desea saber como se podrían relacionar estos, esta metodología es util para determinar dichas relaciones, por ejemplo si tenemos una poblacion de personas de una ciudad, con un conjunto de característicasy deseamos saber como se relacionan estas personas según sus caracteristicas en común con otras personas esta metodología sería útil.

<b><h3>1 ii) Mezcla de Normales</h3>

Descripción:
Modelo de aprendizaje no supervizado para el agrupamiento de datos segun su pertenencia a una distribución normal determinada.


Ventajas:Es más flexible en términos de la covarianza de la agrupación que otros algoritmos de aprendizaje no supervizado, por ejemplo que k-mean, que es un caso específico de GMM.

Desventajas: Si la dimensionalidad del problema es muy grande puede fallar, por razones computacionales.

Ejemplo de aplicaicón:

Si queremos agrupar colores en una imagen, este algoritmo es muy utilizado para la substracción de fondo en video viginalcia, debido a su robustes en cuanto a las variaciones de luminosidad.

Mapas de Kohonen

```r
#SOM example using wines data set
library(kohonen)
library(cluster)
library(ggplot2)
set.seed(7)

#ruta datos
#lee datos del archivo de texto como una matriz. La primera #columna se utiliza como vector de nombres de las filas
datos <-read.table("control2.dat",row.names=1,dec=".",header=TRUE)
datos <- as.data.frame(datos)
summary(datos)
```

```r
#create SOM grid
sommap <- som(scale(datos), grid = somgrid(4, 4, "rectangular"))
plot(sommap)

colors <- function(n, alpha = 1) {
    rev(heat.colors(n, alpha))
}

#plot(sommap, type = "counts", palette.name = colors, heatkey = TRUE)

plot(sommap, type = "mapping", pchs = 20, main = "Mapping Type SOM")

#plot(sommap, type = "dist.neighbours", palette.name = terrain.colors)
```

K-means

```r
#ingresar valor de k0 
k0=6
cl<-kmeans(datos,k0,iter.max=1000,nstart=10)
#grafica los datos coloreados con el grupo asignado en planos de pares de variables
#pairs(datos,col=cl$cluster)
```

```r
#Análisis de Componentes Principales para graficar los grupos obtenidos
#Calcula plano de componentes principales y grafica las observaciones con el número de su grupo
res <- princomp(datos, cor=TRUE)
res2 <- predict(res,cl$centers)
#res2
#cl
biplot(res,xlabs=cl$cluster)
#grafica los centros de cada grupo en el plano de componentes principales utilizando colores distintos
points(res2[,1:2],col=1:k0,pch=19)
```
