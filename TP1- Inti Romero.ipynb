## 3.2 Leer los archivos textos y insertar los datos

```python
# LEYENDO E INSERTANDO A TABLA MEDIOS DE PRENSA
counter = 0
with open("medios.csv") as fichero:
    # recorre l√≠nea a l√≠nea el archivo
    for linea in fichero:
        counter = counter + 1
        # muestra l√≠nea √∫ltima le√≠da
        #print(linea)  
        #PROCESO DE SEPARACION DE CARACTERES
        arraglo = linea.split(',')
        nombre = arraglo[0]
        dueno = arraglo[1]
        follower = arraglo[2]
        #INSERT
        if counter > 1:
            if len(linea)>0:
                print("nombre: "+nombre)
                print("due√±o: "+dueno)
                print("seguidores: "+follower)
                insertarmedios= "INSERT INTO MediosPrensa (pk_nombre, dueno, numfollowers) VALUES ('"+nombre+"','"+dueno+"','"+str(follower)+"')"
                print(insertarmedios)
                cursor = cnx.cursor()
                cursor.execute(insertarmedios)
                cnx.commit()
```

```python
# LEYENDO E INSERTANDO A TABLA NOTICIAS
counter = 0
with open("noticias.txt") as fichero:
    # recorre l√≠nea a l√≠nea el archivo
    for linea in fichero:
        counter = counter + 1
        if counter == 1:
            arraglo = linea.split()
            fecha = arraglo[0]
        if counter == 2:
            arraglo = linea.split()
            foranea = arraglo[0]
        if counter == 3:
            titulo = linea
        #INSERT date: a√±o-mes-dia   xxxx-xx-xx
        if counter == 4:
            texto = linea
            insertarmedios= "INSERT INTO Noticias (titulo, texto, fecha, nombre_medios_prensa) VALUES ('"+titulo+"','"+texto+"','"+fecha+"','"+foranea+"')"
            print("Fecha: "+fecha)
            print("Clave Foranea:--"+foranea+"---")
            print("Largo de foranea: "+str(len(foranea) ))
            print("Titulo: "+titulo)
            print("Noticia: "+texto)
            print("    ")
            cursor = cnx.cursor()
            cursor.execute(insertarmedios)
            cnx.commit()
        if counter == 5:
            counter = 0
        
        
```

# 4. Consultar la base de datos

Listar los medios de prensa en Chile ordenado seg√∫n su n√∫mero de followers en Twitter.

```python
# Prepare SQL query to INSERT a record into the database.
sql = "SELECT pk_nombre , numfollowers FROM MediosPrensa order by numfollowers"


try:
    # Execute the SQL command
    cursor = cnx.cursor()
    cursor.execute(sql)
    # Fetch all the rows in a list of lists.
    results = cursor.fetchall()
    for row in results:
        nombre = row[0]
        followers = row[1]
        # Now print fetched result
        print("Nombre: "+nombre)
        print("Seguidores Twiter: "+str(followers))
        print("  ")
except:
    print ("Error: unable to fetch data")
```

```python
# Prepare SQL query to INSERT a record into the database.
sql = "select count(*) from (SELECT dueno FROM MediosPrensa group by dueno) as subquery"

try:
    # Execute the SQL command
    cursor = cnx.cursor()
    cursor.execute(sql)
    # Fetch all the rows in a list of lists.
    results = cursor.fetchall()
    for row in results:
        numero = row[0]
        # Now print fetched result
        print("Cantidad de due√±os en Chile: "+str(numero))  
except:
    print ("Error: unable to fetch data")
```

Contar cuantes medios de prensa posee cada due√±o.

```python
# Prepare SQL query to INSERT a record into the database.
sql = "SELECT dueno,count(*) FROM MediosPrensa group by dueno"

try:
    # Execute the SQL command
    cursor = cnx.cursor()
    cursor.execute(sql)
    # Fetch all the rows in a list of lists.
    results = cursor.fetchall()
    for row in results:
        dueno = row[0]
        conteo = row[1]
        # Now print fetched result
        print("Due√±o: "+dueno)  
        print("Numero de medios que posee: "+str(conteo))
        print("  ")
except:
    print ("Error: unable to fetch data")
```

Mostrar los √∫ltimos 10 t√≠tulos de noticias del medio "La Tercera"

```python
# Prepare SQL query to INSERT a record into the database.
sql = "SELECT titulo FROM Noticias where nombre_medios_prensa = 'latercera' order by fecha desc"

try:
    # Execute the SQL command
    cursor = cnx.cursor()
    cursor.execute(sql)
    counter = 0
    # Fetch all the rows in a list of lists.
    results = cursor.fetchall()
    for row in results:
        counter = counter + 1
        titulo = row[0]
        if counter < 11:           
            # Now print fetched result
            print("Titulo: "+titulo)  
            print("  ")
        
except:
    print ("Error: unable to fetch data")
```

Mostrar los nombre de medios y los t√≠tulos en los cuales aparacen la palabra "Valdivia".

```python
# Prepare SQL query to INSERT a record into the database.
sql = "SELECT titulo, nombre_medios_prensa FROM Noticias WHERE titulo LIKE '%Valdivia%'"

try:
    # Execute the SQL command
    cursor = cnx.cursor()
    cursor.execute(sql)
    counter = 0
    # Fetch all the rows in a list of lists.
    results = cursor.fetchall()
    for row in results:
        titulo = row[0] 
        medio = row[1]
        # Now print fetched result
        print("Titulo: "+titulo)  
        print("Medio de Prensa: "+medio)  
        print(" --- ")
        
except:
    print ("Error: unable to fetch data")
```

Por cada medio, mostrar el n√∫mero de t√≠tulos en los cuales aparecen la palabra "Puerto Montt".

```python
# Prepare SQL query to INSERT a record into the database.
sql = "SELECT nombre_medios_prensa, count(*) FROM Noticias WHERE titulo LIKE '%Puerto Montt%' group by nombre_medios_prensa"

try:
    # Execute the SQL command
    cursor = cnx.cursor()
    cursor.execute(sql)
    # Fetch all the rows in a list of lists.
    results = cursor.fetchall()
    for row in results:
        medio = row[0] 
        conteo = row[1]
        # Now print fetched result
        print("Medio de Prensa: "+medio)  
        print("Numero de titulos con la palabra: Puerto Montt: "+str(conteo))  
        print(" --- ")
        
except:
    print ("Error: unable to fetch data")
```
